# Segunda Entrega

```python
#Importamos nuestro dataset
data= pd.read_csv("Desktop/Placement_Data_Full_Class.csv")
#Removemos el serial number
data.drop("sl_no",axis=1,inplace= True)
data= data.fillna(0)
data.head()
```

```python
#Preprocesamiento del dataset
data.drop(['ssc_b','hsc_b'], axis=1, inplace=True)
```

### Anexo:
Se optó por eliminar las columnas "ssc_b" y "hsc_b" ya que según los análisis previos no son variables predictoras determinantes para corren nuestro modelo de arbol de decisión.

```python
#Codificamos las variables objeto de nuestro dataset
data["gender"] = data.gender.map({"M":0,"F":1})
data["hsc_s"] = data.hsc_s.map({"Commerce":0,"Science":1,"Arts":2})
data["degree_t"] = data.degree_t.map({"Comm&Mgmt":0,"Sci&Tech":1, "Others":2})
data["workex"] = data.workex.map({"No":0, "Yes":1})
data["status"] = data.status.map({"Not Placed":0, "Placed":1})
data["specialisation"] = data.specialisation.map({"Mkt&HR":0, "Mkt&Fin":1})
```

```python
#Importamos las librerías para realizar nuestro algoritmo de clasificación
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

```python
#Separamos las variables en x e y
X = data[['gender', 'ssc_p', 'hsc_p', 'hsc_s', 'degree_p', 'degree_t', 'workex','etest_p', 'specialisation', 'mba_p',]]
y = data['status']
```

```python
#Separo los datos en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Análisis de Componenetes Principales - PCA

```python
#Normalización de datos
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Aplicación de PCA
from sklearn.decomposition import PCA

pca = PCA()
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
#Análisis de la varianza explicada para cada componente
explained_variance = pca.explained_variance_ratio_
explained_variance
```

### Conclusiones: 
* El primer componente principal es responsable de la varianza del 27,75 %. De igual manera, el segundo
componente causa una variación del 16.46 % y asi sucesivamente.
* En resumen los primeros 6 componentes engloban el 81.37 % de la información de clasificación contenida en el conjunto
de características dentro de la base de datos.

## Algoritmos de Clasificación

### Implementación del algoritmo Árbol de Decisión

```python
#Árbol de decisión
dtree = DecisionTreeClassifier(criterion='entropy')

#Entrenamos el modelo
dtree.fit(X_train, y_train)

#Realizo una prediccion
y_train_pred = dtree.predict(X_train) #Prediccion en Train
y_test_pred = dtree.predict(X_test) #Prediccion en Test
```

```python
#Calculo en accuracy en train
train_accuracy = accuracy_score(y_train, y_train_pred)

#Calculo el acccuracy en test
test_accuracy = accuracy_score(y_test, y_test_pred)

print('% de aciertos sobre el set de entrenamiento:', train_accuracy)
print('% de aciertos sobre el set de evaluación:',test_accuracy)
```

```python
# Estructura del árbol creado
from sklearn.tree import plot_tree
from sklearn.tree import export_graphviz
fig, ax = plt.subplots(figsize=(20,20))

print(f"Profundidad del árbol: {dtree.get_depth()}")
print(f"Número de nodos terminales: {dtree.get_n_leaves()}")

plot = plot_tree(
            decision_tree = dtree,
            feature_names = X.columns,
            class_names   = 'MEDV',
            filled        = True,
            impurity      = False,
            fontsize      = 10,
            precision     = 2,
            ax            = ax
       )
```

```python
importancia_predictores = pd.DataFrame(
                            {'predictor': X.columns,
                             'importancia': dtree.feature_importances_}
                            )
print("Importancia de los predictores en el modelo")
importancia_predictores.sort_values('importancia', ascending=False)
```

```python
#Importancia de las variables

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,8))
importances = dtree.feature_importances_
columns = X.columns
sns.barplot(columns, importances)
plt.title('Importancia de cada Feature',fontsize=15, pad=10)
plt.xlabel("Variables",fontsize=14)
plt.ylabel("Valor",fontsize=14)
plt.tick_params(labelsize=12)
plt.show()
```

```python
#Calculo de indicadores de clasificación (métricas)
print(classification_report(y_test, y_test_pred))
```

### Métricas para la evaluación del modelo

### Precisión: 
* La precisión es la proporción de verdaderos positivos (TP) por la suma de verdaderos positivos (TP) y falsos positivos (FP)


### Recall: 
* El recall es la proporción de verdaderos positivos (TP) por la suma de verdaderos positivos (TP) y falsos negativos (FN)

### F1 Score: 
* La puntuación F1 es la media armónica de precisión y el recall. Se encuentra entre [0,1].
